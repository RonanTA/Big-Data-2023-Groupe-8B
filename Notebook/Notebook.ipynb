## Prétraitrements <a class="anchor" id="chapter2"></a>

### Dictionnaire des données

- **PassengerId** : A unique Id for each passenger. Each Id takes the form gggg_pp where gggg indicates a group the passenger is travelling with and pp is their number within the group. People in a group are often family members, but not always.

- **HomePlanet** : The planet the passenger departed from, typically their planet of permanent residence.

- **CryoSleep** : Indicates whether the passenger elected to be put into suspended animation for the duration of the voyage.Passengers in cryosleep are confined to their cabins.

- **Cabin** : The cabin number where the passenger is staying. Takes the form deck/num/side, where side can be either P for Port or S for Starboard.

- **Destination** : The planet the passenger will be debarking to.

- **Age** : The age of the passenger.

- **VIP** : Whether the passenger has paid for special VIP service during the voyage. 

- **RoomService**, **FoodCourt**, **ShoppingMall**, **Spa**, **VRDeck** : Amount the passenger has billed at each of the Spaceship Titanic's many luxury amenities.

- **Name** : The first and last names of the passenger.

- **Transported** : Whether the passenger was transported to another dimension. This is the target, the column you are trying to predict.

### Imputation par le mode principal

Dans un premier temps, nous allons effectuer une imputation par le mode principal. Cette imputation consiste à affecter le mode à chaque valeur manquante. 

On applique cette méthode pour les variables "Home Planet", "Crysleep", "Destination" et "VIP"

```python
df_mode = df_train.copy()
mode_imputer = SimpleImputer(strategy='most_frequent')
df_mode['HomePlanet'] = mode_imputer.fit_transform(df_mode['HomePlanet'].values.reshape(-1,1))
df_mode['CryoSleep'] = mode_imputer.fit_transform(df_mode['CryoSleep'].values.reshape(-1,1))
df_mode['Destination'] = mode_imputer.fit_transform(df_mode['Destination'].values.reshape(-1,1))
df_mode['VIP'] = mode_imputer.fit_transform(df_mode['VIP'].values.reshape(-1,1))
```

### Suppression des dernières valeurs manquantes

La variable "Cabin" peut être déduite des autres variables en supposant que deux passagers d'un même groupe soient dans la même cabine. Suivant ce principe, il suffit de créer un dictionnaire de données affectant un groupe à une cabine et par la suite effectuer une jointure entre notre base et ce dictionnaire.

```python
import numpy as np

data = np.array(df_mode)
group_list = []
for i in range(df_mode.shape[0]) :
    string = data[i][0]
    short_list = string.split('_')
    tuple_list = []
    tuple_list.append(short_list[0])
    tuple_list.append(data[i][3])
    group_list.append(tuple_list)
    
df_group = pd.DataFrame(group_list)
df_group = df_group.drop_duplicates()
df_group = df_group.dropna(axis=0, how='any', subset=None, inplace=False)
df_group.columns = ['Group','Cabin']

def Add_column(data, processed_list) :
    data = np.array(data)
    massive_list = []
    for i in range(data.shape[0]) :
        liste = []
        for j in range(data.shape[1]) :
            liste.append(data[i][j])
        liste.append(processed_list[i])
        massive_list.append(liste)
    data2 = np.array(massive_list)
    df_maj = pd.DataFrame(data2)
    return df_maj

data = np.array(df_mode)
massive_list_group = []
for i in range(df_mode.shape[0]) :
    string = data[i][0]
    short_list = string.split('_')
    massive_list_group.append(short_list[0])
df_mode = Add_column(df_mode,massive_list_group)

l = []
for i in df_train.columns:
    l.append(i)
l.append('Group')
df_mode.columns = l

df_missing_final = df_mode
```

## Variables de positionnement sur le vaisseau <a class="anchor" id="section_3_1"></a>

### Balisage de la problématique

Deux nouvelles variables peuvent être déduites grâce au numéro de cabine : 
    
- **Deck** : Quai sur le vaisseau
    
- **Side** : Coté du vaisseau

### Implémentation de fonctions utiles

Cette fonction nous permet d'ajouter une colonne au dataset que nous utilisons

```python
def Add_column(data, processed_list) :
    data = np.array(data)
    massive_list = []
    for i in range(data.shape[0]) :
        liste = []
        for j in range(data.shape[1]) :
            liste.append(data[i][j])
        liste.append(processed_list[i])
        massive_list.append(liste)
    data2 = np.array(massive_list)
    df_maj = pd.DataFrame(data2)
    return df_maj
```

```python
df_train_v2.head()
```

### Création des deux nouvelles variables

```python
data = np.array(df_train_v2)
massive_list_deck = []
massive_list_side = []
for i in range(df_train_v2.shape[0]) :
    string = data[i][3]
    if string == 'nan' :
        string = 'X/X/X'
    short_list = string.split('/')
    massive_list_deck.append(short_list[0])
    massive_list_side.append(short_list[2])
dftrain_v3_1 = Add_column(df_train_v2,massive_list_deck)
dftrain_v3_2 = Add_column(dftrain_v3_1,massive_list_side)
```

```python
dftrain_v3_2.head()
```

### Réatribution des colonnes

```python
l = []
for i in df_train_v2.columns:
    l.append(i)
l.append('Deck')
l.append('Side')
    
dftrain_v3_2.columns = l
```
